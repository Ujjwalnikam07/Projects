# Property Price Prediction

Here, We will create a linear model on Property price dataset. It is historical database of properties, based on which we will create a model.

```python
import pandas as pd
```

```python
prop=pd.read_csv("PROPERTYPRICE.CSV")
```

```python

#pd.set_option('display.max_columns',None)
#pd.set_option('display.max_rows',None)
```

```python
prop.iloc[:,0:10].describe(include='all')
```

We will drop unnecessary and redundant variables(coulumns).

```python
prop.drop(["Id","Lane_Type","Fireplace_Quality","Pool_Quality","Fence_Quality","Miscellaneous_Feature"],axis=1,inplace=True)
```

```python
prop.shape
```

```python
prop.describe(include='all')
```

Too many nulls, Now we will deal with the nulls in batch of 20 variables at a time.

```python
prop.iloc[:,0:20].isnull().sum()
```

```python
prop.iloc[:,20:50].isnull().sum()
```

```python
prop.drop(['Lot_Extent',          
'Basement_Height',              
'Exposure_Level',               
'BsmtFinType2',                 
'Garage_Built_Year',          
'Garage_Finish_Year',         
'Garage_Condition' ],inplace=True,axis=1)
```

```python
prop.shape
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
prop.isnull().sum()
```

```python
prop.describe(include='all')
```

```python
prop.Basement_Condition=prop.Basement_Condition.fillna('TA')
prop.BsmtFinType1=prop.BsmtFinType1.fillna('Unf')
prop.Garage=prop.Garage.fillna('Attchd')
prop.Garage_Quality=prop.Garage_Quality.fillna('TA')
```

```python
prop.isnull().sum()
```

```python
prop.shape
```

```python
prop.dropna(inplace=True)
```

```python
prop.shape
```

Now that the data is cleaned, we will convert the object variables into numerical ones.

```python
import sklearn
from sklearn.preprocessing  import LabelEncoder
le=LabelEncoder()
```

```python
x=prop
for i in range(1,x.shape[1]):
    if x.dtypes[i]=='object':
        x[x.columns[i]]=le.fit_transform(x[x.columns[i]])
```

```python
prop.describe(include='all')
```

```python
prop.shape
```

The real task begins from here, we will split the data into x and y for preparing to create a model.

```python
prop_x=prop.iloc[:,0:67]
prop_y=prop.iloc[:,-1]
```

```python
prop_y.head()
```

```python
from sklearn.model_selection import train_test_split
prop_x_train,prop_x_test,prop_y_train,prop_y_test=train_test_split(prop_x,prop_y,test_size=.2,random_state=101)
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
reg.fit(prop_x_train,prop_y_train)
```

```python
prop_rsq=reg.score(prop_x_train,prop_y_train)
prop_rsq
```

```python
prop_adjrsq=1-(((1-prop_rsq)*((prop_x_train.shape[0]-1)))/((prop_x_train.shape[0])-prop_x_train.shape[1]-1))
prop_adjrsq
```

We have got adjusted r square of around 83%

```python
prop_predict=reg.predict(prop_x_test)
```

```python
from sklearn import metrics
```

```python
prop_mse=metrics.mean_squared_error(prop_y_test,prop_predict)
prop_mse
```

```python
prop_mae=metrics.median_absolute_error(prop_y_test,prop_predict)
prop_mae
```

```python
import numpy as np
```

```python
prop_rmse=np.sqrt(prop_mse)
prop_rmse
```

```python
error=prop_y_test-prop_predict
per_err=error/prop_y_test
abs_err=np.absolute(per_err)
prop_mape=np.mean(abs_err)
prop_mape
```

Error Percentage in our model is around 10%.
